```python
for col in data.columns:
  data.rename(columns={col:col.lower().replace(' ','_')},inplace=True)
data.columns
# correcting the column names
```

```python
data["date"] = pd.to_datetime(data.date)
#fixxing date format
```

```python
data["amount"] = data.amount.str.replace('[$,]','',regex=True).astype(float)
#fixing amount column to make it in float
```
